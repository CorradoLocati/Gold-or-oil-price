```python
# Create the gold stock price dataframe by reading and loading gold_stocks_price.csv. gsp = gold stock price
gsp = pd.read_csv(r'C:\Users\bocco\OneDrive\Documents\Education\20240708-20250208 - LSE - Data Analytics Career Accelerator'
           r'\Course 2\Module 2\Data\gold_stocks_price.csv')  # gsp = gold stock price

print(gsp.columns)
print(gsp.shape)
print(gsp.dtypes)
gsp.head()
```

```python
# Create the oil price dataframe by reading and loading oil_price.csv. op = oil price
op = pd.read_csv(r'C:\Users\bocco\OneDrive\Documents\Education\20240708-20250208 - LSE - Data Analytics Career Accelerator'
           r'\Course 2\Module 2\Data\oil_price.csv')

print(op.columns)
print(op.shape)
print(op.dtypes)
op.head()
```

```python
# Create a subset of the gsp dataframe with only the 'Date', 'Open', 'High', 'Low' columns

gsp_subset = gsp[['Date', 'Open', 'High', 'Low']]
print(gsp_subset.columns)
print(gsp_subset.shape)
print(gsp_subset.dtypes)
gsp_subset.head()
```

```python
# Create a subset of the previous subset with only the first 500 rows

gsp_subset_500 = gsp_subset.iloc[:500,:]
print(gsp_subset_500.columns)
print(gsp_subset_500.shape)
print(gsp_subset_500.dtypes)
print(gsp_subset_500.isna().sum()) # Prints the number of missing values per column
gsp_subset_500.head()
```

```python
# Create a subset of the op dataframe with only the 'Date', 'Open', 'High', 'Low' columns, and the first 500 rows

op_subset_500 = op[['Date', 'Open', 'High', 'Low']].iloc[:500]
print(op_subset_500.columns)
print(op_subset_500.shape)
print(op_subset_500.dtypes)
print(op_subset_500.isna().sum()) # Prints the number of missing values per column
op_subset_500.head()
```
